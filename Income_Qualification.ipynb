# Count how many null values are existing in columns.

```python
for i in income_df:
    print(f"Null Values {i} in column is --> {income_df[i].isnull().sum()/income_df.shape[0] * 100}")
```

```python
for i in income_df:
    if (income_df[i].isnull().sum()/income_df.shape[0] * 100) > 50 :
        income_df.drop(i, axis=1, inplace=True)
    elif (income_df[i].isnull().sum()/income_df.shape[0] * 100) > 0 and income_df[i].dtypes == 'float64':
        income_df.fillna(income_df[i].mean(), inplace=True)
```

```python
for i in income_df:
    print(f"Null Values {i} in column is --> {income_df[i].isnull().sum()/income_df.shape[0] * 100}")
```
