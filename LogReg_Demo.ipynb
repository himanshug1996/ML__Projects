```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
```

```python
bc_bunch = load_breast_cancer()

print(bc_bunch.DESCR)
```

```python
cancer_df = pd.DataFrame(bc_bunch.data ,columns =bc_bunch.feature_names)
cancer_df
```

```python
cancer_df["Class"] = bc_bunch.target

cancer_df.head()
```

```python
cancer_df["Class"].value_counts().plot.bar()
```

we infer that 

 - 212 malignant data points are there
 - 357 benign data points are there

```python
X = cancer_df.drop("Class" ,axis =1)
y = cancer_df["Class"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X , y ,test_size = 0.3, random_state =21)
```

```python
print("Shape of X_train is : ", X_train.shape)
print("Shape of X_test is : ", X_test.shape)
print("Shape of y_train is : ", y_train.shape)
print("Shape of y_test is : ", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg_model = LogisticRegression(solver = "newton-cg") 
# --> When your data is continuous numerical 'newton-cg' is more appropriate than 'lbfgs'.
# --> newton-cg -- within max_iter=100 there is still convergence so we put newton-cg to increase interations.
```

```python
logreg_model.fit(X_train,y_train)
```

```python
y_pred_test = logreg_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print("The accuracy score of the model on test data is :")
print(accuracy_score(y_test,y_pred_test)) # --> Actual values & Predicted value
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred_test))
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test,y_pred_test))
```

### Thus in binary classification, 
the count of true negatives is : math:`C_{0,0}`
-false negatives is : math:`C_{1,0}
-true positives is math:`C_{1,1}` 
-false positives is :math:`C_{0,1}
                              


```python
y_pred_train = logreg_model.predict(X_train)
```

```python
print("The accuracy score of the model on train data is :")
print(accuracy_score(y_train,y_pred_train)) # --> Actual values & Predicted value
```
